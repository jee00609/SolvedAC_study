```python
## 틀린 코드

answer = 0
n = int(input())
lost = list(map(int,input().split()))
reserve = list(map(int,input().split()))

lost.sort()
reserve.sort()
    
data_list = [0 for _ in range(n)]

for i in lost:
    if i in reserve:
        data_list[i-1]=0
    else:
        data_list[i-1]=-1

for i in reserve:
    if i in lost:
        data_list[i-1]=0
    else:
        data_list[i-1]=1
        
#if 순서를 고치면 통과다
for i in range(len(data_list)):        
    if i!=n-1 and data_list[i]==-1 and data_list[i+1] == 1:
        data_list[i] = 0
        data_list[i+1] = 0
        
    elif i!=0 and data_list[i]==-1 and data_list[i-1]==1:
        data_list[i] = 0
        data_list[i-1] = 0   

for i in data_list:
    if i == 0 or i == 1:
        answer+=1
        
print(data_list)
print(answer)
```

```python
#체육복
#성공 코드
def solution(n, lost, reserve):
    answer = 0
    
    lost.sort()
    reserve.sort()

    data_list = [0 for _ in range(n)]

    for i in lost:
        if i in reserve:
            data_list[i-1]=0
        else:
            data_list[i-1]=-1

    for i in reserve:
        if i in lost:
            data_list[i-1]=0
        else:
            data_list[i-1]=1


    for i in range(len(data_list)):        
        if i!=0 and data_list[i]==-1 and data_list[i-1]==1:
            data_list[i] = 0
            data_list[i-1] = 0  
        
        if i!=n-1 and data_list[i]==-1 and data_list[i+1] == 1:
            data_list[i] = 0
            data_list[i+1] = 0


    for i in data_list:
        if i == 0 or i == 1:
            answer+=1
    
    
    return answer
```

```python
#큰 수 만들기
#시간 초과 뜬다
#당연하지!
import time
from itertools import combinations, permutations

start_time = time.time()

def solution(number, k):
    number = list(map(''.join, combinations(number, len(number)-k)))
    answer = max(number)
    return answer


print(solution("1924",2))
print(solution("1231234",3))
print(solution("4177252841",4))

end_time= time.time()
print("time = ",end_time-start_time)
```

```python
#https://kdgt-programmer.tistory.com/5?category=1121942
# 참고
# 난 멍청이야.....

def solution(number, k):
    num = list(number) 
    stack = [num[0]]
    #빼지는 개수, count == k 면 종료
    count = 0

    for i in range(1, len(num)):
        if count == k: 
            stack = stack + num[i:] 
            break
        stack.append(num[i])
        
        if stack[-1] > stack[-2]:
            while(len(stack) != 1 and stack[-1] > stack[-2] and count < k):
                #순서 바꾸기
                stack[-1],stack[-2] = stack[-2],stack[-1]
                stack.pop()
                #뺀 개수 증가
                count+=1
    answer = "".join(stack[:len(num)-k])
    return answer

print(solution("1924",2))
print(solution("1231234",3))
print(solution("4177252841",4))
#stack 을 바로 꺼내면 안되는 이유
print(solution("99991", 3))
```
