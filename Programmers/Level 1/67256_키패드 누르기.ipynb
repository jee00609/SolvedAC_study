이 코드에선 실제 좌표값에 대한 것이 아닌 움직인 값을 이야기하고 있다.

따라서 4와3에 대해서 각각 5 까지 움직인 값을 구할 때 선의 개수로 생각해보면 쉽게 풀 수 있었다.

4와 5는 작대기 하나

3과 5는 작대기 두개

즉 x,y 값의 피타고라스를 이용한 좌표 값이 아닌 그냥 x끼리 차이의 값 + y끼리의 차이의 값이었다.

```python
## 성공 코드
## 문제를 쉽게 생각했다.

import math
def solution(numbers, hand):
    
    answer = ''
    key = {'1':(0,0),'2':(0,1),'3':(0,2),'4':(1,0),'5':(1,1),'6':(1,2),'7':(2,0),'8':(2,1),'9':(2,2),'*':(3,0),'0':(3,1),'#':(3,2)}
    left_hand = key['*']
    right_hand = key['#']
    
    for i in numbers:
        if i == 1 or i == 4 or i ==7:
            answer+='L'
            left_hand = key[str(i)]

        elif i==3 or i == 6 or i ==9:
            answer+='R'
            right_hand = key[str(i)]
            
        else:
            left_x = left_hand[0]-key[str(i)][0]
            left_y = left_hand[1]-key[str(i)][1]
            left = math.fabs(left_x)+math.fabs(left_y)

            right_x = right_hand[0]-key[str(i)][0]
            right_y = right_hand[1]-key[str(i)][1]
            right = math.fabs(right_x)+math.fabs(right_y)
            
            
            
            if left == right:
                if hand == "right":
                    answer+='R'
                    right_hand = key[str(i)]
                else:
                    answer+='L'
                    left_hand = key[str(i)]
                    
            else:
                if left < right:
                    answer+='L'
                    left_hand = key[str(i)]
                else:
                    answer+='R'
                    right_hand = key[str(i)]

        
    return answer
```

```python
## 추천 코드
## 나와 코드가 비슷하지만 좀 더 효율적이다.
## 막 dfs bfs 쓰는 문제일 줄 알았는데 다행이다.


def solution(numbers, hand):
    answer = ''
    key_dict = {1:(0,0),2:(0,1),3:(0,2),
                4:(1,0),5:(1,1),6:(1,2),
                7:(2,0),8:(2,1),9:(2,2),
                '*':(3,0),0:(3,1),'#':(3,2)}

    left = [1,4,7]
    right = [3,6,9]
    lhand = '*'
    rhand = '#'
    for i in numbers:
        if i in left:
            answer += 'L'
            lhand = i
        elif i in right:
            answer += 'R'
            rhand = i
        else:
            curPos = key_dict[i]
            lPos = key_dict[lhand]
            rPos = key_dict[rhand]
            ldist = abs(curPos[0]-lPos[0]) + abs(curPos[1]-lPos[1])
            rdist = abs(curPos[0]-rPos[0]) + abs(curPos[1]-rPos[1])

            if ldist < rdist:
                answer += 'L'
                lhand = i
            elif ldist > rdist:
                answer += 'R'
                rhand = i
            else:
                if hand == 'left':
                    answer += 'L'
                    lhand = i
                else:
                    answer += 'R'
                    rhand = i

    return answer
```
