```python
## 성공

from collections import Counter

def solution(N, stages):
    
    # 총 사용자 수 user
    user = len(stages)
    print("user is ",user)
    
    # stages 를 dict 로 변환
    stages_dict = Counter(stages)
    dict(stages_dict)
    
    # 모든 유저가 성공해서 dict 요소에 빠진 스테이지를 인위로 채워줌
    for i in range(1,N+1):
        if i not in stages_dict:
            stages_dict[i]=0
    
    # 스테이지 dict 를 오름차순 정렬
    stages_dict = dict(sorted(stages_dict.items(), key = lambda item: item[0], reverse = False))
    print(stages_dict)
    
    
    # 핵심 코드
    fail = {}
    for i in range(1,N+1):
        temp = 0
        print("i is ",i)
        
        # 내 코드에선 dict 1 의 요소에 대해서 user = user-temp 를 제대로 못해줘서
        # 인위적으로 만들어줌
        if i == 1 :
            fail[1]=stages_dict[i]/user
        else:
            # 여기를 1,i+1 이 아닌 1,i 로 변경
            for k in range(1,i):
                temp += stages_dict[k]
            # 만약 value 가 0 일 경우 division 오류를 예외 처리 하기 위해 미리 0.0 으로 세팅    
            if stages_dict[i] == 0:
                fail[i]=0.0
            else:
                # user - temp 가 0 이 될 경우 (즉 모든 유저가 해당 문제에 정체되있는 경우) 를 예외처리 하기 위해 세팅
                if (user-temp)==0:
                    fail[i]=1.0
                else:
                    fail[i] = stages_dict[i] / (user-temp)
                print("fail is ",fail[i])

    print("fail is ",fail)
    
    #실패율을 내림차순 정렬
    sort_fail = dict(sorted(fail.items(), key = lambda item: item[1], reverse = True))
    print("sort fail is ",sort_fail)
    
    # 정렬된 실패율 dict 의 key 값만 추출
    answer = list(sort_fail.keys())
    return answer
```

```python
# 추천 코드 1
# 어떻게 이렇게 푸는 거지....
# 다만 list.count()가 O(n)이라 시간복잡도가 O(n^2)

def solution(N, stages):
    result = {}
    denominator = len(stages)
    for stage in range(1, N+1):
        if denominator != 0:
            count = stages.count(stage)
            result[stage] = count / denominator
            denominator -= count
        else:
            result[stage] = 0
    return sorted(result, key=lambda x : result[x], reverse=True)
```

```python
# 추천 코드 2

def solution(N, stages):
    fail = {}
    for i in range(1,N+1):
        try:
            fail_ = len([a for a in stages if a==i])/len([a for a in stages if a>=i])
        except:
            fail_ = 0
        fail[i]=fail_
    answer = sorted(fail, key=fail.get, reverse=True)
    return answer
```

```python
# 추천 코드 3
# 시간 복잡도 효율성이 좋다

def solution(N, stages):
    answer = []
    fail = []
    info = [0] * (N + 2)
    for stage in stages:
        info[stage] += 1
    for i in range(N):
        be = sum(info[(i + 1):])
        yet = info[i + 1]
        if be == 0:
            fail.append((str(i + 1), 0))
        else:
            fail.append((str(i + 1), yet / be))
    for item in sorted(fail, key=lambda x: x[1], reverse=True):
        answer.append(int(item[0]))
    return answer
```
