```python
# 뭔가 더 쉽게 풀 수 있는 문제같은데 말이다...
# 이건 뭔가 아름답지 못한 코드일 것 같다.
# 뭔가 split 이나 정규식으로 푸는 문제같은데
# 그리고 느낌상 큐 같은 걸로 풀 것 같은? 느낌이?

import math

def solution(dartResult):
    
    temp = list(dartResult)
    myData = []
    answer = 0
    
    
    #가장 먼저 S D T * # 을 기준으로 나눠준다.
    for i in range(len(temp)):
        
        # S 는 1의 제곱이니 숫자를 그대로 저장
        if temp[i] == 'S':
            # 이때 10 의 정수에 대해서 따로 처리해준다.
            # 이는 10 이 두자리 이기 떄문!
            if i>1 and temp[i-1]=='0' and temp[i-2]=='1':
                dart = 10
            else:
                dart = int(temp[i-1])
            myData.append(dart)
        
        # D 는 2의 제곱이니 pow 2 제곱으로 처리
        elif temp[i] == 'D':
            if i>1 and temp[i-1]=='0' and temp[i-2]=='1':
                dart = int(math.pow(10,2))
            else:
                dart = int(math.pow(int(temp[i-1]),2))
            myData.append(dart)
        
        # T 는 3 의 제곱이니 pow 3의 제곱으로 처리
        elif temp[i] == 'T':
            if i>1 and temp[i-1]=='0' and temp[i-2]=='1':
                dart = int(math.pow(10,3))
            else:
                dart = int(math.pow(int(temp[i-1]),3))
            myData.append(dart)
            
        elif temp[i] == '*':
            myData.append(temp[i])
        
        elif temp[i] == '#':
            myData.append(temp[i])
            
    # * # 을 처리해주자
    for i in range(len(myData)):
        
        # 두배 곱셈일 경우
        if myData[i] == '*':
            count = 0
            for k in range(i-1,-1,-1):
                # 최대 두개의 정수를 2배할 수 있기에 count 를 처리해준다.
                # 또한 * # 이 아닌 정수만을 취급해준다
                if count <2 and type(myData[k])==int:
                    myData[k] = myData[k]*2
                    count +=1
        # -1 곱셈일 경우            
        if myData[i] == '#':
            myData[i-1] *= -1
    
    # 이후 * # 을 제외한 모든 정수를 더해준다.
    for i in myData:
        if type(i) == int:
            answer += i
            
    return answer
```

```python
# 추천 코드 1
# 정규식을 사용해서 엄청 아름답게 풀었다.
# 이런 코드를 짜고파!

import re


def solution(dartResult):
    bonus = {'S' : 1, 'D' : 2, 'T' : 3}
    option = {'' : 1, '*' : 2, '#' : -1}
    p = re.compile('(\d+)([SDT])([*#]?)')
    dart = p.findall(dartResult)
    for i in range(len(dart)):
        if dart[i][2] == '*' and i > 0:
            dart[i-1] *= 2
        dart[i] = int(dart[i][0]) ** bonus[dart[i][1]] * option[dart[i][2]]

    answer = sum(dart)
    return answer
```

```python
# 추천 코드 2

def solution(dartResult):
    point = []
    answer = []
    dartResult = dartResult.replace('10','k')
    point = ['10' if i == 'k' else i for i in dartResult]
    print(point)

    i = -1
    sdt = ['S', 'D', 'T']
    for j in point:
        if j in sdt :
            answer[i] = answer[i] ** (sdt.index(j)+1)
        elif j == '*':
            answer[i] = answer[i] * 2
            if i != 0 :
                answer[i - 1] = answer[i - 1] * 2
        elif j == '#':
            answer[i] = answer[i] * (-1)
        else:
            answer.append(int(j))
            i += 1
    return sum(answer)
```
