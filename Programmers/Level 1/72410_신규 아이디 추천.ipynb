## 현재 풀지 못함

틀린 케이스 : 7,12,16,24,26

```python
import re

def solution(new_id):
    
    #1단계
    new_id = new_id.lower()
    print(new_id)
    
    #2단계
    stay_characters = '0123456789abcdefghijklmnopqrstuvwxyz-_.'
    new_id = ''.join( x for x in new_id if x in stay_characters)
    print("2단계 ",new_id)
    
    #3단계
    list_new_id = []
    list_new_id.append(new_id[0])
    for i in range(1,len(new_id)):
        if new_id[i] != new_id[i-1]:
            list_new_id.append(new_id[i])
            
    #new_id = ''.join(list_new_id)
    print("3단계 ",list_new_id)
    
    #4단계
    if len(list_new_id)!=0 and list_new_id[0] == '.':
        list_new_id = list_new_id[1:]
    
    if len(list_new_id)!=0 and list_new_id[-1] == '.':
        list_new_id = list_new_id[:-1]
    print("4단계 ",list_new_id)
    
    #5단계
    if len(list_new_id) == 0:
        list_new_id.append('a')
    print("5단계 ",list_new_id)
    
    #6단계
    if len(list_new_id) > 15:
        list_new_id = list_new_id[:15]
        #만약 마지막이 . 이라면 삭제!
        if list_new_id[-1] == '.':
            list_new_id = list_new_id[:-1]
    print("6단계 ",list_new_id)
    
    
    #7단계
    i = len(list_new_id)
    if i <3:
        while i < 3:
            list_new_id.append(list_new_id[-1])
            i = len(list_new_id)
    print("7단계 ",list_new_id)  
    
    answer = ''.join(list_new_id)
    return answer
```
