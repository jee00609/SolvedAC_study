```python
#DFS 기초
#깊이 우선 탐색

# 혼자서도 외워서 쓸 수 있도록 공부하자

import time

start_time = time.time()

def dfs(graph, v, visited):
    #현재 노드를 방문 처리
    visited[v] = True
#     print(v,end=' ')
    
    for i in graph[v]:
        
        if not visited[i]:
            print("visited[i] i is ",i)
            print(graph[v])
            dfs(graph,i,visited)
        
graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]   

print(graph)

#각 노드가 방문된 정보를 리스트 자료형으로 표현
visited = [False]*9
print(visited)

dfs(graph,1,visited)

end_time = time.time()
print("Time = ",end_time-start_time)
```

```python
#BFS 기초
#너비 우선 탐색
#큐
#일반적인 경우 DFS 보다 실제 수행 시간이 좋다.

# 혼자서도 외워서 쓸 수 있도록 공부하자

import time
#queue 를 위해
from collections import deque

start_time = time.time()

def bfs(graph,start,visited):
    #큐 구현을 위한 dequeue
    queue = deque([start])
    visited[start] = True
    while queue:
        v = queue.popleft()
        print(v,end=' ')
        
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True

                
graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]
visited = [False]*9

bfs(graph,1,visited)

end_time = time.time()
print("")
print("Time = ",end_time-start_time)
```

```python
import time
#이동 과정에 가중치 혹은 제약이 있을땐 dfs 를 이용하자

start_time = time.time()

n,m = map(int,input().split())

# graph = list(map(int,input().split()))
# print(graph)

graph = []

for i in range(n):
    graph.append(list(map(int,input())))
    
def dfs(x,y):
    #주어진 범위를 벗어나면 즉시 종료
    if x<= -1 or y<=-1 or x >=n or y>=m:
        return False
    
    #방문 안했던 좌표면
    if graph[x][y] == 0:
        #방문 처리
        graph[x][y] = 1
        #상하좌우 재귀 호출
        dfs(x-1,y)
        dfs(x+1,y)
        dfs(x,y-1)
        dfs(x,y+1)
        return True
    return False

count = 0

for i in range(n):
    for j in range(m):
        if dfs(i,j)==True:
            count+=1
            
print(count)
        
end_time = time.time()
print("Time = ",end_time-start_time)
```

```python
#미로탈출

import time
from collections import deque


start_time = time.time()

n,m = map(int,input().split())
graph = []

dx = [-1,1,0,0]
dy = [0,0,-1,1]

#split 붙여서 list Error 뜸! 붙이지마!
for _ in range(n):
    graph.append(list(map(int,input())))

    
def bfs(x,y):
    queue = deque()
    queue.append((x,y))
    
    while queue:
        x,y = queue.popleft()
        for i in range(4):
            nx = x+dx[i]
            ny = y+dy[i]
            
            if nx < 0 or ny < 0 or nx >= n or ny >=m:
                continue
            
            if graph[nx][ny] == 0:
                continue
        
            if graph[nx][ny] == 1:
                graph[nx][ny] = graph[x][y]+1
                queue.append((nx,ny))
                
    return graph[n-1][m-1]

print(bfs(0,0))

end_time = time.time()
print("Time = ",end_time-start_time)
```
