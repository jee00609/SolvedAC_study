```python
#구현_4_1

import time
start_time = time.time()

n = int(input())
x,y = 1,1

plans = input().split()

# L R U D
dx = [0,0,-1,1]
dy = [-1,1,0,0]
moves_type = ['L','R','U','D']


for plan in plans:
    for i in range(len(moves_type)):
        if plan == moves_type[i]:
            nx = x + dx[i]
            ny = y + dy[i]

    if nx < 1 or ny<1 or nx > n or ny > n:
        continue
        
    x,y = nx,ny
    
print("result = ",x,",",y)

end_time =  time.time()
print("Time : ",end_time-start_time)
```

```python
#구현_4_2
#왕실 나이트
import time
start_time = time.time()

#a1
#col,row
input_data = input()
row = int(input_data[1])
#시작이 1,1 ~ 8,8
col = int(ord(input_data[0]))- int(ord('a'))+1

move_type = [(-1,2),(-1,-2),(1,2),(1,-2),(-2,1),(-2,-1),(2,1),(2,-1)]
count = 0

for move in move_type:
    nr = row + move[0]
    nc = col + move[1]
    
    if nr >= 1 and nr <=8 and nc >= 1 and nc <=8:
        count+=1
    
print(count)

end_time =  time.time()
print("Time : ",end_time-start_time)
```
