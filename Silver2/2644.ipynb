```python
## 성공
## family 의 리스트 값을  [] 가 만들어 지지 않도록 맞춤 제작

def bfs(a,b):
    total = 0
    queue = [[a, total]]
    
    while queue:
        value = queue.pop(0)
        v = value[0]
        count = value[1]
        
        if v == b:
            return count
        
        if visited[v-1]==False:
            visited[v-1]=True
            count+=1
            for i in family[v-1]:
                if not visited[i-1]:
                    queue.append([i, count]) # 각 value 값마다 count 가 올라가게 만들어 여러번 세지 않도록 만들었다.
            
    return -1
    
n = int(input())
person1,person2 = map(int, input().split())
m = int(input())

family = [[] for _ in range(n)]
visited = [False for _ in range(n)]

for _ in range(m):
    x,y = map(int, input().split())
    family[x-1].append(y)
    family[y-1].append(x)


result = bfs(person1,person2)
print(result)
```
