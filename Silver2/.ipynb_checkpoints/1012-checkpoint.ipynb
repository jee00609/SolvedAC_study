```python
## 런타임 에러
## 어디서 틀린 걸까?

def dfs(x,y):
    dx = [0,0,-1,1]
    dy = [-1,1,0,0]
    
    visit[x][y] = 1
    
    for i in range(4):
        nowX = x + dx[i]
        nowY = y + dy[i]
        
        if 0<=nowX and 0<=nowY and nowX < N and nowY < M: 
            if matrix[nowX][nowY] == 1 and visit[nowX][nowY] == 0:
                visit[nowX][nowY] = 1
                dfs(nowX,nowY)


TestCase = int(input())

result = []

for _ in range(TestCase):
    M,N,K = map(int, input().split())
    matrix = [[0]*M for i in range(N)]
    visit = [[0]*M for i in range(N)]
    larva = 0
    for _ in range(K):
        x,y = map(int, input().split())
        matrix[y][x] = 1
    
    for i in range(N):
        for j in range(M):
            if matrix[i][j] == 1 and visit[i][j] == 0:
                dfs(i,j)
                larva = larva+1
    result.append(larva)
    
    
for i in result:
    print(i)
```

```python
## 입력값에 제한을 준 코드
## 또 런타임 에러가 발생했다 돌아버릴것같아

def dfs(x,y):
    dx = [0,0,-1,1]
    dy = [-1,1,0,0]
    
    visit[x][y] = 1
    
    for i in range(4):
        nowX = x + dx[i]
        nowY = y + dy[i]
        
        if 0<=nowX and 0<=nowY and nowX < N and nowY < M: 
            if matrix[nowX][nowY] == 1 and visit[nowX][nowY] == 0:
                visit[nowX][nowY] = 1
                dfs(nowX,nowY)


TestCase = int(input())

result = []

for _ in range(TestCase):
    M,N,K = map(int, input().split())
    matrix = [[0]*M for i in range(N)]
    visit = [[0]*M for i in range(N)]
    larva = 0
    for _ in range(K):
        col,row = map(int, input().split())
        if 0 < row and 0 < col and row < N and col < M:
            matrix[row][col] = 1
    
    for i in range(N):
        for j in range(M):
            if matrix[i][j] == 1 and visit[i][j] == 0:
                dfs(i,j)
                larva = larva+1
    result.append(larva)
    
    
for i in result:
    print(i)
```

```python
## 이번엔 런타임 에러가 아니라 "틀립니다"가 뜸
## 돌아버릴 것 같아 ㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋ

import sys
sys.setrecursionlimit(10**6)

def dfs(x,y):
    dx = [0,0,-1,1]
    dy = [-1,1,0,0]
    
    visit[x][y] = 1
    
    for i in range(4):
        nowX = x + dx[i]
        nowY = y + dy[i]
        
        # 왼 오 위 아래 좌표들이 행렬 안의 값일 경우
        if 0<=nowX and 0<=nowY and nowX < N and nowY < M:
            
            #배추가 있으며 , 방문하지 않았으면
            if matrix[nowX][nowY] == 1 and visit[nowX][nowY] == 0:
                visit[nowX][nowY] = 1
                dfs(nowX,nowY)


TestCase = int(input())

result = [] #각 테스트 케이스마다 얻을 정답들을 모아 둘 장소

for _ in range(TestCase):
    M,N,K = map(int, input().split())
    matrix = [[0]*M for i in range(N)]
    visit = [[0]*M for i in range(N)]
    larva = 0
    
    for _ in range(K):
        col,row = map(int, input().split())
        if 0 < row and 0 < col and row < N and col < M:
            matrix[row][col] = 1
    
    for i in range(N):
        for j in range(M):
            ## 배추가 있으며, 방문을 안했을 경우
            if matrix[i][j] == 1 and visit[i][j] == 0:
                dfs(i,j)
                larva = larva+1
                
    result.append(larva)
    

# 값들 출력
for i in result:
    print(i)
```
