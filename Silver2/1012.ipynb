```python
## 좀 더 코드 밥꿔봄
## 성공
## 31680kb 332ms
import sys
sys.setrecursionlimit(10**6)

def dfs(x,y,cnt):
    dx, dy = [-1, 1, 0, 0], [0, 0, -1, 1]
    
    visit[x][y] = cnt
    
    for i in range(4):
        nowX = x + dx[i]
        nowY = y + dy[i]
        
        # 왼 오 위 아래 좌표들이 행렬 안의 값일 경우
        if 0<=nowX and 0<=nowY and nowX < N and nowY < M:
            
            #배추가 있으며 , 방문하지 않았으면
            if matrix[nowX][nowY] == 1 and visit[nowX][nowY] == -1:
                dfs(nowX,nowY,cnt)


TestCase = int(input())

result = [] #각 테스트 케이스마다 얻을 정답들을 모아 둘 장소

for _ in range(TestCase):
    M,N,K = map(int, input().split())
    matrix = [[0]*M for _ in range(N)]
    visit = [[-1]*M for _ in range(N)]
    #애벌레
    larva = 0
    
    for _ in range(K):
        # 실제론 이렇게 짜면 오류가 뜨기 쉽지만, 제대로 입력 값을 넣을 거란 전제하에 if 문 삭제
        col,row = map(int, input().split())
        matrix[row][col] = 1
    
    for i in range(N):
        for j in range(M):
            ## 배추가 있으며, 방문을 안했을 경우
            if matrix[i][j] == 1 and visit[i][j] == -1:
                larva = larva+1 
                dfs(i,j,larva) #애벌레 값을 넣어보자
                
    result.append(larva)
    

# 값들 출력
for i in result:
    print(i)
```

```python
## 29076kb 76ms
## jiyolla 님이 작성한 코드
## 이런걸 생각해낼 수 있다는 게 대단해

import sys
read = sys.stdin.readline
n_test = int(read())

for _ in range(n_test):
    m, n, k = map(int, read().split())
    field = [([0] * (m+2)) for _ in range(n + 2)]
    for _ in range(k):
        x, y = map(int, read().split())
        field[y + 1][x + 1] = 1

    stack = []
    n_worm = 0
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if field[i][j] == 1:
                n_worm += 1
                stack.append((j, i))
                while stack:
                    x, y = stack.pop()
                    if field[y][x + 1] == 1:
                        field[y][x + 1] = 0
                        stack.append((x + 1, y))
                    if field[y][x - 1] == 1:
                        field[y][x - 1] = 0
                        stack.append((x - 1, y))
                    if field[y + 1][x] == 1:
                        field[y + 1][x] = 0
                        stack.append((x, y + 1))
                    if field[y - 1][x] == 1:
                        field[y - 1][x] = 0
                        stack.append((x, y - 1))
    print(n_worm)
```
