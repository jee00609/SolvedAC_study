```python
## 시간 초과
TestCase = int(input())

ResultList = []

for i in range(TestCase):
    a, b = map(int, input().split())
    UserInput = a**b
    result = UserInput%10
    
    if result == 0:
        result = 10
    
    ResultList.append(result)
    
for i in range(TestCase):
    print(ResultList[i])
```

시간 초과로 인한 에러를 해결하기 위해 각 숫자 제곱 값에 대한 반복 숫자를 찾아보았다.

1 -> 1
2 -> 2,4,6,8
3 -> 3,9,7,1
4 -> 4,6
5 -> 5
6 -> 6
7 -> 7,9,3,1
8 -> 8,4,2,6
9 -> 9,1
10 -> 0

11 이후로는 

11 -> 1
12 -> 2,4,6,8

과 같이 첫번째 자리의 숫자에 따라 반복되는 값이 같아졌다.

사용자의 입력 a 값이 1,5,6,10 일 경우 if 값으로 그냥 바로 ResultList 에 정해진 값을 입력하는게 좋을 듯 싶다.

만약 리스트를 짠다면

TwoList = [2,4,8,6]

이 될것이고 

사용자의 b 입력 값에 대한 계산식을 짜게 된다면

(b%4)-1 을 해야 할 듯 싶다.

```python
# 틀렸습니다! 
# 왜 틀렸지 

TestCase = int(input())

TwoList = [2,4,6,8]
ThreeList = [3,9,7,1]
FourList = [4,6,4,6]
SevenList = [7,9,3,1]
EightList = [8,4,2,6]
NineList = [9,1,9,1]

ResultList = []

for _ in range(TestCase):
    a, b = map(int, input().split())
    
    UserInput = a % 10
    
    if UserInput == 1:
        result = 1
    if UserInput == 5:
        result = 5
    if UserInput == 6:
        result = 6
    if UserInput == 0:
        result = 10
        
    else :
        listNum = (b%4)-1
        
        if UserInput == 2:
            result = TwoList[listNum]
        if UserInput == 3:
            result = ThreeList[listNum]
        if UserInput == 4:
            result = FourList[listNum]
        if UserInput == 7:
            result = SevenList[listNum]
        if UserInput == 8:
            result = EightList[listNum]
        if UserInput == 9:
            result = NineList[listNum]
    
    ResultList.append(result)
    
for i in range(TestCase):
    print(ResultList[i])
```
